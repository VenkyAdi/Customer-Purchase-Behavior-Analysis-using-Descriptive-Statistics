```python
columns_with_outliers = ['Income', 'Age', 'MntWines', 'MntFruits', 'MntMeatProducts']

def remove_outliers(data, columns):

    for column in columns:
        Q1 = data[column].quantile(0.20)
        Q3 = data[column].quantile(0.80)
        IQR = Q3 - Q1


        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR


        data = data[(data[column] >= lower_limit) & (data[column] <= upper_limit)]

    return data


df_new = remove_outliers(cs_df, columns_with_outliers)
```

```python
d_variables = ['Income', 'Age', 'Recency']

num_variables = len(d_variables)

fig, axes = plt.subplots(nrows=  2, ncols=2, figsize=(12, 7))

axes = axes.flatten()

for i, variable in enumerate(d_variables):
    sns.boxplot(x=variable, data=df_new, ax=axes[i],color = '#9370DB')
    axes[i].set_title(f'Boxplot of {variable}')

plt.tight_layout()
plt.show()
```

# **Probability and Expected Value**

```python
import numpy as np
from scipy.stats import norm
```

```python
m_recency = np.mean(df_new['Recency'])
std_recency = np.std(df_new['Recency'])

prob_less_20 = norm.cdf(20, loc = m_recency, scale = std_recency)
prob_grt_40 = 1- norm.cdf(40,loc = m_recency , scale = std_recency)

print(f"Probability of recency less than 20: {prob_less_20:.2f}")
print(f"Probability of recency greater than 40: {prob_grt_40:.2f}")

expected_recency = m_recency
print(f"The expected_recency is: {expected_recency:.2f}")

plt.figure(figsize=(8, 6))
plt.boxplot(df_new['Recency'],patch_artist=True, boxprops=dict(facecolor="#9370DB", color="#9370DB"))
plt.title('Box plot of Recency')
plt.xlabel('Recency')
plt.ylabel('Values')
plt.axhline(y=expected_recency, color='black', linestyle='--', label='Expected Recency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
m_sal = np.mean(df_new['Income'])
std_sal = np.std(df_new['Income'])

prob_less_40000 = norm.cdf(40000, loc = m_sal, scale = std_sal)
prob_grt_60000 = 1- norm.cdf(60000,loc = m_sal , scale = std_sal)

print(f"Probability of income less than 40000$: {prob_less_40000:.3f}")
print(f"Probability of income greater than 60000$: {prob_grt_60000:.3f}")

expected_salary = m_sal
print(f"The expected income is: {expected_salary:.2f}")

plt.figure(figsize=(8, 6))
plt.boxplot(df_new['Income'],patch_artist=True, boxprops=dict(facecolor="#9370DB", color="#9370DB"))
plt.title('Box plot of Income')
plt.xlabel('Income')
plt.ylabel('Income')
plt.axhline(y=expected_salary, color='black', linestyle='--', label='Expected Income')
plt.legend()
plt.grid(True)
plt.show()
```

```python
m_age = np.mean(df_new['Age'])
std_age = np.std(df_new['Age'])

prob_less_30 = norm.cdf(30, loc = m_age, scale = std_age)
prob_grt_50 = 1- norm.cdf(50,loc = m_age , scale = std_age)

print(f"Probability of age less than 30: {prob_less_30:.2f}")
print(f"Probability of age greater than 50: {prob_grt_50:.2f}")

expected_age = m_age
print(f"The expected age is: {expected_age:.2f}")

plt.figure(figsize=(8, 6))
plt.boxplot(df_new['Age'],patch_artist=True, boxprops=dict(facecolor="#9370DB", color="#9370DB"))
plt.title('Box plot of Age')
plt.xlabel('Age')
plt.ylabel('Age')
plt.axhline(y=expected_age, color='black', linestyle='--', label='Expected Age')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Bio_var = ['Response','Complain']

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12,8))
fig.suptitle("Binomial Distribution ", y = 0.98, color = 'black')

axes = axes.flatten()

for i , var in enumerate(Bio_var):
  sns.histplot(df_new[var], kde = True, ax = axes[i], color = '#9370DB')
  axes[i].set_title(f'Distribution of {var}:')

for i in range(len(Bio_var), len(axes)):
  fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import binom

num_response = len(cs_df['Response'])
success_response = np.mean(cs_df['Response'])
k1 = np.sum(cs_df['Response'] == 1)


num_Complain = len(cs_df['Complain'])
success_Complain = np.mean(cs_df['Complain'])
k2 = np.sum(cs_df['Complain'] == 1)



prob_response = binom.pmf(k1,num_response,success_response)
prob_complain = binom.pmf(k2,num_Complain,success_Complain)

print(f"The probability of Response being 1 : {prob_response:.2f}")
print(f"The probability of Complain being 1 : {prob_complain:.2f}")



expected_response = num_response * success_response
expected_complain = num_Complain * success_Complain

print(f"The Expected value of Response is : {expected_response:.2f}")
print(f"The Expected value of Complain is: {expected_complain:.2f}")
```
